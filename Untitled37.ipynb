```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
L = 1.0           # Longitud del dominio espacial
T = 0.5           # Tiempo final
h_values = [0.1]  # Tamaño de paso espacial (fijo)
k_values = [0.0005, 0.01]  # Tamaños de paso temporal

# Solución exacta
def exact_solution(x, t):
    return np.exp(-np.pi**2 * t) * np.sin(np.pi * x)

# Configuración de la malla espacial
x = np.linspace(0, L, int(L/h_values[0]) + 1)
Nx = len(x) - 1

# Preparar tablas y gráficos
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

for idx, k in enumerate(k_values):
    h = h_values[0]
    Nt = int(T / k)  # Número de pasos temporales
    r = k / h**2     # Parámetro de estabilidad

    # Verificar estabilidad
    if r > 0.5:
        print(f'Advertencia: k={k}, h={h} produce r={r:.3f} (>0.5). El método puede ser inestable.')

    # Inicializar solución numérica
    w = np.zeros((Nx + 1, Nt + 1))
    w[:, 0] = np.sin(np.pi * x)  # Condición inicial

    # Iteración temporal
    for j in range(Nt):
        for i in range(1, Nx):
            w[i, j+1] = w[i, j] + r * (w[i+1, j] - 2*w[i, j] + w[i-1, j])

    # Solución exacta en t=0.5
    u_exact = exact_solution(x, T)

    # Comparación tabular
    print(f"\nComparación para k={k}:")
    print("x_i\tu(x_i,0.5)\t\tw_i\t\tError")
    for i in range(len(x)):
        error = abs(u_exact[i] - w[i, -1])
        print(f"{x[i]:.1f}\t{u_exact[i]:.8f}\t{w[i, -1]:.8f}\t{error:.2e}")

    # Gráfica comparativa
    axes[idx].plot(x, u_exact, 'b-', label='Solución Exacta')
    axes[idx].plot(x, w[:, -1], 'ro--', label='Solución Numérica')
    axes[idx].set_title(f'k = {k}')
    axes[idx].set_xlabel('x')
    axes[idx].set_ylabel('u(x, 0.5)')
    axes[idx].legend()
    axes[idx].grid()

plt.tight_layout()
plt.show()

# Análisis de diferencias
print("\n--- Análisis de resultados ---")
print("Para k=0.0005 (r=0.05): El método es estable y preciso.")
print("Para k=0.01 (r=2.0): La violación del criterio de estabilidad (r>0.5) causa oscilaciones no físicas.")
```
