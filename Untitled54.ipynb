# Método de Galerkin (EDO con condiciones de frontera)

Este es un método de **residuos pesados**, donde las funciones de peso coinciden con las funciones base.

Resolveremos el problema:

$
-u''(x)+u(x)-x=0,\qquad 0\le x\le 1,
$
$
u(0)=u(1)=0.
$

Usaremos una aproximación con 3 funciones base:

$
u_3(x)=\sum_{i=1}^{3}\alpha_i\,\beta_i(x),\qquad \beta_i(x)=\sin(i\pi x).
$

> Nota de errata: si en el texto aparece \(\phi\) en lugar de \(\beta\), se refiere a las funciones base.

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

##  Solución exacta

Primero obtenemos la solución exacta con `sympy` para compararla con la aproximación numérica.



```python
x = sp.symbols('x', real=True)
u = sp.Function('u')

# EDO del enunciado: -u'' + u - x = 0
ode = -sp.diff(u(x), x, 2) + u(x) - x

sol_ex = sp.dsolve(ode, ics={u(0): 0, u(1): 0})
u_ex_expr = sp.simplify(sol_ex.rhs)

print("Solución exacta u(x):")
sp.pprint(u_ex_expr)

# Versión numérica para graficar
u_ex = sp.lambdify(x, u_ex_expr, "numpy")
```

## Aproximación de Galerkin con senos

Tomamos:

$
u_3(x)=\alpha_1\sin(\pi x)+\alpha_2\sin(2\pi x)+\alpha_3\sin(3\pi x).
$

Estas funciones cumplen automáticamente:

-$ \sin(i\pi\cdot 0)=0\Rightarrow u_3(0)=0$

-$ \sin(i\pi\cdot 1)=\sin(i\pi)=0\Rightarrow u_3(1)=0$

Así que la condición de frontera queda satisfecha “gratis”.

```python
pi = sp.pi

# Coeficientes desconocidos
a1, a2, a3 = sp.symbols('a1 a2 a3', real=True)
alpha = sp.Matrix([a1, a2, a3])

# Funciones base beta_i = sin(i*pi*x)
beta = [sp.sin(1*pi*x), sp.sin(2*pi*x), sp.sin(3*pi*x)]

# Construimos u3(x)
u3 = a1*beta[0] + a2*beta[1] + a3*beta[2]
print("u3(x) =")
sp.pprint(u3)
```

## Residuo (error) y ecuaciones de Galerkin

Definimos el **residuo** al sustituir \(u_3\) en la EDO:

$
R(x) = -u_3''(x) + u_3(x) - x.
$

Galerkin impone que el residuo sea ortogonal a cada función de peso (aquí, las mismas $\beta_j$):

$
\int_0^1 R(x)\,\beta_j(x)\,dx = 0,\qquad j=1,2,3.
$

Esto produce un sistema lineal para $\alpha_1,\alpha_2,\alpha_3$.

```python
# Residuo
R = -sp.diff(u3, x, 2) + u3 - x
R = sp.simplify(R)

print("Residuo R(x) =")
sp.pprint(R)

# Ecuaciones Galerkin: integral(R*beta_j)=0
eqs = [sp.Eq(sp.integrate(R*beta[j], (x, 0, 1)), 0) for j in range(3)]

print("\nEcuaciones de Galerkin:")
for e in eqs:
    sp.pprint(sp.simplify(e))
```

##  Forma matricial A·α = b

Vamos a construir el sistema:

$
A_{j i}=\int_0^1\left(-\beta_i''(x)+\beta_i(x)\right)\beta_j(x)\,dx,\qquad
b_j=\int_0^1 x\,\beta_j(x)\,dx.
$

De modo que:

$
\sum_{i=1}^3 A_{j i}\alpha_i = b_j.
$

```python
# Construcción de A y b de forma explícita
A = sp.zeros(3,3)
b = sp.zeros(3,1)

for j in range(3):
    # b_j = ∫ x beta_j dx
    b[j,0] = sp.integrate(x*beta[j], (x,0,1))
    for i in range(3):
        A[j,i] = sp.integrate((-sp.diff(beta[i], x, 2) + beta[i]) * beta[j], (x,0,1))

A_s = sp.simplify(A)
b_s = sp.simplify(b)

print("Matriz A =")
sp.pprint(A_s)

print("\nVector b =")
sp.pprint(b_s)

# Resolución del sistema
alpha_sol = sp.simplify(A_s.LUsolve(b_s))
print("\nCoeficientes alpha =")
sp.pprint(alpha_sol)
```

##  u₃(x) final (sustituyendo α)

Sustituimos los coeficientes y obtenemos la aproximación cerrada $u_3(x)$

```python
u3_final = sp.simplify(u3.subs({a1: alpha_sol[0], a2: alpha_sol[1], a3: alpha_sol[2]}))

print("u3(x) aproximada (Galerkin N=3) =")
sp.pprint(u3_final)

# Función numérica
u3_num = sp.lambdify(x, u3_final, "numpy")
```

##  Comparación gráfica: exacta vs Galerkin

```python
xx = np.linspace(0, 1, 600)
y_exact = u_ex(xx)
y_gal = u3_num(xx)

plt.figure(figsize=(9,5))
plt.plot(xx, y_exact, label="Solución exacta", linewidth=2)
plt.plot(xx, y_gal, "--", label="Galerkin (N=3)", linewidth=2)
plt.title("Comparación: Solución exacta vs Galerkin (senos)")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.grid(True)
plt.legend()
plt.show()
```

## Error de la aproximación

Definimos el error puntual:
$
e(x)=u(x)-u_3(x).
$

Calcularemos:
- Error máximo aproximado: $\|e\|_\infty \approx \max|e(x)|$ en una malla fina.
- RMSE discreto: $\sqrt{\frac{1}{N}\sum e(x_k)^2}$

```python
err = y_exact - y_gal

err_inf = np.max(np.abs(err))
rmse = np.sqrt(np.mean(err**2))

print(f"Error máximo (aprox ||e||_inf): {err_inf:.12e}")
print(f"RMSE (error L2 discreto):       {rmse:.12e}")

plt.figure(figsize=(9,4))
plt.plot(xx, err, linewidth=2)
plt.title("Error puntual e(x) = u(x) - u3(x)")
plt.xlabel("x")
plt.ylabel("e(x)")
plt.grid(True)
plt.show()
```

La gráfica representa el error puntual
$
e(x)=u(x)-u_3(x)
$
en $0\le x\le 1$. Si $e(x)\approx 0$, la aproximación es buena; si $e(x)>0$, $u_3$ subestima a $u$; si $e(x)<0$, $u_3$ sobreestima a $u$. Además, $\|e\|_\infty\approx\max|e(x)|$ es el error máximo y el RMSE es una medida del error promedio cuadrático en la malla.
