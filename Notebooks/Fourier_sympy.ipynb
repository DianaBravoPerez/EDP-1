```python
from sympy import fourier_series, pi, plot, Piecewise
from sympy.abc import x
```

```python
f=Piecewise(
    (0, x < -2),
    (x+2,(x >= -2) & (x < 0)),
    (0,(x<=2)))
```

```python
plot(f, (x,-2,2),line_color ='red')
```

```python
s = fourier_series(f, (x, -2, 2))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
from sympy import fourier_series, pi, plot, Piecewise, exp, symbols, And
from sympy.abc import x

# Problema 1: Serie de Fourier en cosenos de e^(2x) en [0,1]
# Definición de la función f1, una función definida por partes
# Para x en [0,1], f1 = e^(2x), y para x en [-1,0), f1 = e^(-2x), y en el resto es 0
f1 = Piecewise(
    (exp(2*x), And(x >= 0, x <= 1)),
    (exp(-2*x), And(x >= -1, x < 0)),
    (0, True)
)
# Cálculo de la serie de Fourier de la función f1 en el intervalo [-1,1]
s1 = fourier_series(f1, (x, -1, 1))
# Truncamiento de la serie de Fourier a n=3, n=5 y n=7 términos
s1_3 = s1.truncate(n=3)
s1_5 = s1.truncate(n=5)
s1_7 = s1.truncate(n=7)
# Graficar la función original y las series truncadas para diferentes valores de n
p1 = plot(f1, s1_3, s1_5, s1_7, (x, 0, 1), show=False, legend=True)
p1[0].label = 'Original'  # Etiqueta para la función original
p1[1].label = 'n=3'  # Etiqueta para la serie truncada a 3 términos
p1[2].label = 'n=5'  # Etiqueta para la serie truncada a 5 términos
p1[3].label = 'n=7'  # Etiqueta para la serie truncada a 7 términos
# Mostrar la gráfica con las etiquetas
p1.show()

# Problema 2: Serie de Fourier en senos de e^(2x) en [0,1]
# Definición de la función f2, otra función definida por partes
# Para x en [0,1], f2 = e^(2x), y para x en [-1,0), f2 = -e^(-2x), y en el resto es 0
f2 = Piecewise(
    (exp(2*x), And(x >= 0, x <= 1)),
    (-exp(-2*x), And(x >= -1, x < 0)),
    (0, True)
)
# Cálculo de la serie de Fourier de la función f2 en el intervalo [-1,1]
s2 = fourier_series(f2, (x, -1, 1))
# Truncamiento de la serie de Fourier a n=3, n=5 y n=7 términos
s2_3 = s2.truncate(n=3)
s2_5 = s2.truncate(n=5)
s2_7 = s2.truncate(n=7)
# Graficar la función original y las series truncadas para diferentes valores de n
p2 = plot(f2, s2_3, s2_5, s2_7, (x, 0, 1), show=False, legend=True)
p2[0].label = 'Original'  # Etiqueta para la función original
p2[1].label = 'n=3'  # Etiqueta para la serie truncada a 3 términos
p2[2].label = 'n=5'  # Etiqueta para la serie truncada a 5 términos
p2[3].label = 'n=7'  # Etiqueta para la serie truncada a 7 términos
# Mostrar la gráfica con las etiquetas
p2.show()

# Problema 3: Extensión par (cosenos)
# Definición de la función f3_cos, función periódica definida por partes
# La función tiene valores 1 y 2 en diferentes intervalos dentro de [0, pi]
f3_cos = Piecewise(
    (1, And(x >= 0, x <= pi/2)),
    (2, And(x > pi/2, x <= pi)),
    (1, And(x >= -pi/2, x < 0)),
    (2, And(x >= -pi, x < -pi/2)),
    (0, True)
)
# Cálculo de la serie de Fourier de la función f3_cos en el intervalo [-pi, pi]
s3_cos = fourier_series(f3_cos, (x, -pi, pi))
# Truncamiento de la serie de Fourier a n=3, n=5 y n=7 términos
s3_cos_3 = s3_cos.truncate(n=3)
s3_cos_5 = s3_cos.truncate(n=5)
s3_cos_7 = s3_cos.truncate(n=7)
# Graficar la función original y las series truncadas para diferentes valores de n
p3_cos = plot(f3_cos, s3_cos_3, s3_cos_5, s3_cos_7, (x, 0, pi), show=False, legend=True)
p3_cos[0].label = 'Original'  # Etiqueta para la función original
p3_cos[1].label = 'n=3'  # Etiqueta para la serie truncada a 3 términos
p3_cos[2].label = 'n=5'  # Etiqueta para la serie truncada a 5 términos
p3_cos[3].label = 'n=7'  # Etiqueta para la serie truncada a 7 términos
# Mostrar la gráfica con las etiquetas
p3_cos.show()

# Problema 3: Extensión impar (senos)
# Definición de la función f3_sin, función periódica definida por partes
# La función tiene valores 1, 2, -1 y -2 en diferentes intervalos dentro de [0, pi]
f3_sin = Piecewise(
    (1, And(x >= 0, x <= pi/2)),
    (2, And(x > pi/2, x <= pi)),
    (-1, And(x >= -pi/2, x < 0)),
    (-2, And(x >= -pi, x < -pi/2)),
    (0, True)
)
# Cálculo de la serie de Fourier de la función f3_sin en el intervalo [-pi, pi]
s3_sin = fourier_series(f3_sin, (x, -pi, pi))
# Truncamiento de la serie de Fourier a n=3, n=5 y n=7 términos
s3_sin_3 = s3_sin.truncate(n=3)
s3_sin_5 = s3_sin.truncate(n=5)
s3_sin_7 = s3_sin.truncate(n=7)
# Graficar la función original y las series truncadas para diferentes valores de n
p3_sin = plot(f3_sin, s3_sin_3, s3_sin_5, s3_sin_7, (x, 0, pi), show=False, legend=True)
p3_sin[0].label = 'Original'  # Etiqueta para la función original
p3_sin[1].label = 'n=3'  # Etiqueta para la serie truncada a 3 términos
p3_sin[2].label = 'n=5'  # Etiqueta para la serie truncada a 5 términos
p3_sin[3].label = 'n=7'  # Etiqueta para la serie truncada a 7 términos
# Mostrar la gráfica con las etiquetas
p3_sin.show()
```
