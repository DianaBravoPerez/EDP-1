ejercicio 2: escriba una función que simule un volado

yo

```python
def volado():
    if r.random() <= 0.5:
        return "sol"
    else:
        return "águila"

print(volado())
```

```python
num_simu = 100
resultados_simulacion = []

for _ in range(num_simu):
    resultados_simulacion.append(volado())

print(f"Se realizaron {len(resultados_simulacion)} simulaciones de volado.")
```

```python
conteo_sol = 0
proporciones_acumuladas_sol = []

for i, resultado in enumerate(resultados_simulacion):
    if resultado == 'sol':
        conteo_sol += 1
    proporcion_actual = conteo_sol / (i + 1)
    proporciones_acumuladas_sol.append(proporcion_actual)

print(f"Se calcularon {len(proporciones_acumuladas_sol)} proporciones acumuladas.")
print(f"Las primeras 10 proporciones acumuladas de 'sol' son: {proporciones_acumuladas_sol[:10]}")
```

```python
import matplotlib.pyplot as plt

# Crear un rango de números para el eje X que represente el número de volados
x_axis = range(1, num_simu + 1) # Se corrige para usar num_simu en lugar de num_simulaciones

# Generar el gráfico de línea
plt.figure(figsize=(12, 6))
plt.plot(x_axis, proporciones_acumuladas_sol, label='Proporción acumulada de "Sol"')

# Añadir una línea horizontal en y=0.5
plt.axhline(y=0.5, color='r', linestyle='--', label='Probabilidad teórica (0.5)')

# Establecer el título y las etiquetas de los ejes
plt.title('Convergencia de la Proporción Acumulada de "Sol"')
plt.xlabel('Número de Volados')
plt.ylabel('Proporción Acumulada de "Sol"')

# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
