```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# ============================================================
# MÉTODO DE MÍNIMOS CUADRADOS – Ejercicio 5.8
#
#   u''(x) + u(x) = x
#   u(0) = 0,   u'(0) = 2
#
# Aproximamos con:
#   u_N(x) = a0 + a1*x + a2*x**2 + a3*x**3
#
# y usamos el método de mínimos cuadrados (en el intervalo [0, 1])
# para encontrar los coeficientes. Luego comparamos con la solución
# exacta y calculamos el error de la aproximación numérica.
# ============================================================

# 1) Se importan las librerías:
#    - sympy: cálculo simbólico
#    - numpy: arreglos numéricos
#    - matplotlib: gráficas

# Variable independiente y función incógnita
x = sp.symbols('x')
u = sp.Function('u')

# 2) Planteamos la EDO:
#       u''(x) + u(x) = x
ode = sp.Eq(sp.diff(u(x), x, 2) + u(x), x)

# 3) Resolvemos la EDO y aplicamos condiciones iniciales:
#       u(0) = 0,  u'(0) = 2
sol_exacta = sp.dsolve(ode, ics={u(0): 0, sp.diff(u(x), x).subs(x, 0): 2})
print("Solución exacta simbólica u(x):")
print(sol_exacta)

# Extraemos la expresión de u(x)
u_exacta_expr = sol_exacta.rhs   # u(x) = x + sin(x)

# 4) Propuesta de solución aproximada:
#       u_N(x) = a0 + a1 x + a2 x^2 + a3 x^3
a0, a1, a2, a3 = sp.symbols('a0 a1 a2 a3')
uN = a0 + a1*x + a2*x**2 + a3*x**3

# 5) Imponemos las condiciones iniciales sobre el polinomio
uN_dx = sp.diff(uN, x)                # u_N'(x)

ecuacion_CI1 = sp.Eq(uN.subs(x, 0), 0)     # u_N(0) = 0
ecuacion_CI2 = sp.Eq(uN_dx.subs(x, 0), 2)  # u_N'(0) = 2

print("\nCondiciones iniciales aplicadas al polinomio:")
print("  CI1:", ecuacion_CI1)
print("  CI2:", ecuacion_CI2)

sol_CI = sp.solve((ecuacion_CI1, ecuacion_CI2), (a0, a1))
a0_val = sol_CI[a0]
a1_val = sol_CI[a1]
print("\nValores de a0 y a1 que satisfacen las CI:")
print("  a0 =", a0_val)
print("  a1 =", a1_val)

# Sustituimos a0 y a1. u_N queda sólo en función de a2 y a3
uN_reducida = uN.subs({a0: a0_val, a1: a1_val})
print("\nPolinomio aproximado después de imponer CI:")
print("  u_N(x) =", uN_reducida)

# 6) Definimos el residuo (error general de la EDO):
#       R(x) = u_N''(x) + u_N(x) - x
#   Esta es la F(x) que se minimiza en el método de mínimos cuadrados.
uN_reducida_dx2 = sp.diff(uN_reducida, x, 2)
R = uN_reducida_dx2 + uN_reducida - x
R_simplificada = sp.simplify(R)

print("\nResiduo R(x) = u_N'' + u_N - x (error de la EDO):")
print("  R(x) =", R_simplificada)  # Es un polinomio cúbico en x

# 7) MÉTODO DE MÍNIMOS CUADRADOS (continuo)
#    Elegimos el intervalo [0, 1].
#    Buscamos a2 y a3 que minimicen
#       J(a2, a3) = ∫_0^1 [R(x)]^2 dx
#    (Si en tu clase usan otro intervalo, cambias [0,1] por el que toque).
a2_sym, a3_sym = sp.symbols('a2 a3')
R_para_J = R_simplificada.subs({a2: a2_sym, a3: a3_sym})

J = sp.integrate(R_para_J**2, (x, 0, 1))
J = sp.simplify(J)

print("\nFuncional de error J(a2, a3) = ∫_0^1 R(x)^2 dx:")
print("  J(a2,a3) =", J)

# 8) Ecuaciones normales de mínimos cuadrados:
#       ∂J/∂a2 = 0,   ∂J/∂a3 = 0
dJ_da2 = sp.diff(J, a2_sym)
dJ_da3 = sp.diff(J, a3_sym)

print("\nEcuaciones normales:")
print("  dJ/da2 =", dJ_da2)
print("  dJ/da3 =", dJ_da3)

sol_a2_a3 = sp.solve((sp.Eq(dJ_da2, 0), sp.Eq(dJ_da3, 0)), (a2_sym, a3_sym))
a2_val = sp.simplify(sol_a2_a3[a2_sym])
a3_val = sp.simplify(sol_a2_a3[a3_sym])

print("\nValores óptimos (mínimos cuadrados) de a2 y a3:")
print("  a2 =", a2_val)
print("  a3 =", a3_val)

# 9) Solución aproximada final u_N(x)
uN_aprox = sp.simplify(uN_reducida.subs({a2: a2_val, a3: a3_val}))
print("\nSolución aproximada u_N(x):")
print("  u_N(x) =", uN_aprox)

# 10) Error puntual:
#       e(x) = u_exacta(x) - u_N(x)
e_expr = sp.simplify(u_exacta_expr - uN_aprox)
print("\nError simbólico e(x) = u(x) - u_N(x):")
print("  e(x) =", e_expr)

# Error global tipo norma L2 en [0,1]:
error_L2_cuadrado = sp.integrate(e_expr**2, (x, 0, 1))
error_L2 = sp.sqrt(error_L2_cuadrado)
print("\nError L2 en [0,1]: ||e||_2 =", error_L2)

# 11) Pasamos a funciones numéricas para graficar
u_exacta_num = sp.lambdify(x, u_exacta_expr, "numpy")
uN_num = sp.lambdify(x, uN_aprox, "numpy")
e_num = sp.lambdify(x, e_expr, "numpy")

# Malla de puntos en [0,1] para comparar numéricamente
xx = np.linspace(0, 1, 200)
yy_exacta = u_exacta_num(xx)
yy_aprox = uN_num(xx)
yy_error = e_num(xx)

# Error máximo aproximado (norma infinito en la malla)
error_max = np.max(np.abs(yy_error))
print("\nError máximo aproximado en la malla de [0,1]:")
print("  ||e||_∞ ≈", float(error_max))

# 12) Gráficas
plt.figure(figsize=(8, 5))
plt.plot(xx, yy_exacta, label='Solución exacta u(x) = x + sin(x)')
plt.plot(xx, yy_aprox, '--', label='Aprox. mínimos cuadrados u_N(x)')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Comparación: solución exacta vs aproximación por mínimos cuadrados')
plt.legend()
plt.grid(True)

plt.figure(figsize=(8, 4))
plt.plot(xx, yy_error, label='Error e(x) = u(x) - u_N(x)')
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Error de la aproximación en [0,1]')
plt.legend()
plt.grid(True)

plt.show()
```
