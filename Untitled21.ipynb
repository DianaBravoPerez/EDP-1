**a)  Graficar en un solo panel los polinomios de Legendre, $ P_n(x)  para n = 0,..,5 $



```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

# Valores de x para la gráfica
x = np.linspace(-1, 1, 500)

# Graficar los polinomios de Legendre para n = 0 a 5
for n in range(6):
    Pn = legendre(n)
    y = Pn(x)
    plt.plot(x, y, label=f'$P_{n}(x)$')

# Configurar la gráfica
plt.xlabel('x')
plt.ylabel('$P_n(x)$')
plt.title('Polinomios de Legendre')
plt.grid(True)
plt.legend()
plt.show()
```

**b) MOSTRAR GRAFICAMENTE QUE SE SATISFACE LA RELACIÓN PARA (n=3):
$$\[ (1 - x^2)P_3'(x) = (3 + 1)xP_3(x) - (3 + 1)P_{4}(x) \]
\[ (1 - x^2)\frac{dP_3(x)}{dx} = 4xP_3(x) - 4P_4(x) \]$$**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import legendre, symbols, diff

# Definir el símbolo x
x = symbols('x')

# Configuración del gráfico
plt.figure(figsize=(10, 6))
plt.title('Verificación de la relación para polinomios de Legendre (n=3)')
plt.xlabel('x')
plt.grid(True)

# Definir n=3
n = 3

# Calcular los polinomios necesarios
P3 = legendre(n, x)
P4 = legendre(n+1, x)
dP3 = diff(P3, x)  # Derivada de P3

# Lado izquierdo de la ecuación: (1-x²)P3'(x)
lhs = (1 - x**2)*dP3

# Lado derecho de la ecuación: (n+1)xP3(x) - (n+1)P4(x)
rhs = (n+1)*x*P3 - (n+1)*P4

# Convertir a funciones numéricas
lhs_func = lambda x_val: lhs.subs(x, x_val)
rhs_func = lambda x_val: rhs.subs(x, x_val)

# Evaluar en el intervalo [-1, 1]
x_vals = np.linspace(-1, 1, 400)
lhs_vals = [lhs_func(val) for val in x_vals]
rhs_vals = [rhs_func(val) for val in x_vals]

# Graficar ambos lados
plt.plot(x_vals, lhs_vals, label='Lado izquierdo: $(1-x^2)P_3\'(x)$', linewidth=2)
plt.plot(x_vals, rhs_vals, '--', label='Lado derecho: $4xP_3(x) - 4P_4(x)$', linewidth=2)

plt.legend(loc='best')
plt.ylabel('Valor de la expresión')
plt.show()
```

**MOSTRAR ALGEBRAICAMENTE QUE SE SATISFACE LA RELACION PARA n=3**
$$\[ (1 - x^2)\frac{d}{dx}\left(\frac{5x^3 - 3x}{2}\right) = 4x\left(\frac{5x^3 - 3x}{2}\right) - 4\left(\frac{35x^4 - 30x^2 + 3}{8}\right) \]$$






```python
from sympy import symbols, diff, legendre, simplify, Eq, latex

# Definir la variable simbólica
x = symbols('x')

# Definir n=3
n = 3

# Obtener los polinomios de Legendre P₃ y P₄
P3 = legendre(n, x)
P4 = legendre(n+1, x)

# Calcular la derivada P₃'
P3_derivada = diff(P3, x)

# Calcular el lado izquierdo: (1-x²)P₃'(x)
lado_izquierdo = (1 - x**2) * P3_derivada

# Calcular el lado derecho: (n+1)xP₃(x) - (n+1)P₄(x) = 4xP₃(x) - 4P₄(x)
lado_derecho = (n + 1)*x*P3 - (n + 1)*P4

# Simplificar ambos lados
lado_izquierdo_simp = simplify(lado_izquierdo)
lado_derecho_simp = simplify(lado_derecho)

# Verificar si son iguales
son_iguales = simplify(lado_izquierdo - lado_derecho) == 0

# Mostrar los resultados
print("Polinomio P₃(x):")
print(latex(P3))
print("\nPolinomio P₄(x):")
print(latex(P4))
print("\nDerivada P₃'(x):")
print(latex(P3_derivada))
print("\nLado izquierdo (1-x²)P₃'(x):")
print(latex(lado_izquierdo_simp))
print("\nLado derecho 4xP₃(x) - 4P₄(x):")
print(latex(lado_derecho_simp))
print("\n¿Son iguales ambos lados?", son_iguales)

```

d) **MOSTRAR ALGEBRAICAMENTE QUE SE SATISFACE LA RELACION PARA n=3
$$\[ (3 + 1)P_{4}(x) - (2\cdot3 + 1)xP_3(x) + 3P_{2}(x) = 0 \]
\[ 4P_4(x) - 7xP_3(x) + 3P_2(x) = 0 \]$$**

```python
from sympy import symbols, legendre, simplify

x = symbols('x')
n = 3

# Polinomios necesarios
P2 = legendre(2, x)
P3 = legendre(3, x)
P4 = legendre(4, x)

# Aplicar la relación de recurrencia
relacion = (n+1)*P4 - (2*n+1)*x*P3 + n*P2

# Simplificar
resultado = simplify(relacion)

print("Resultado de la relación de recurrencia para n=3:")
print(resultado)
```

**Muestre que se cumple la formula de Rodrigues
$$\[ P_n(x) = \frac{1}{2^n n!} \frac{d^n}{dx^n}(x^2 - 1)^n \]$$**

```python
from sympy import symbols, diff, factorial, simplify, expand, Eq
from sympy import legendre as P  # Polinomios de Legendre estándar

x = symbols('x')

def rodrigues(n):
    """polinomio usando la fórmula de Rodrigues"""
    return (1/(2**n * factorial(n))) * diff((x**2 - 1)**n, x, n)

# Verificación para n de 0 a 5
for n in range(6):
    legendre_std = P(n, x)

    legendre_rodrigues = rodrigues(n)

    # Simplificamos para comparar
    rodrigues_simp = simplify(expand(legendre_rodrigues))

    # Verificamos si son iguales
    iguales = simplify(legendre_std - rodrigues_simp) == 0

    print(f"\nn = {n}")
    print(f"Polinomio estándar: {legendre_std}")
    print(f"Fórmula de Rodrigues: {rodrigues_simp}")
    print(f"¿Coinciden? {iguales}")
```
