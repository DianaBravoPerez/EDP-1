### **Ecuación de onda**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Parámetros del problema
L = 1      # Longitud de la cuerda
k = 1      # Altura del triángulo
c = 1      # Velocidad de propagación
N = 20     # Número de términos en la serie
t = 0.2    # Tiempo para evaluar la solución

# Condición inicial triangular
def f(x):
    if 0 <= x <= L/2:
        return (2 * k / L) * x
    elif L/2 < x <= L:
        return (2 * k / L) * (L - x)
    else:
        return 0

# Coeficientes A_n
def An(n):
    integrand = lambda x: f(x) * np.sin(n * np.pi * x / L)
    result, _ = quad(integrand, 0, L)
    return (2 / L) * result

# Solución u(x, t)
def u(x, t):
    suma = 0
    for n in range(1, N+1):
        if n % 2 == 1:  # Solo n impares aportan
            a_n = An(n)
            suma += a_n * np.cos(n * np.pi * c * t / L) * np.sin(n * np.pi * x / L)
    return suma

# Crear arreglo de posiciones y calcular u(x,t)
x_vals = np.linspace(0, L, 500)
u_vals = [u(x, t) for x in x_vals]

# Graficar
plt.plot(x_vals, u_vals, label=f't = {t}')
plt.title('Solución de la ecuación de onda con deformación triangular')
plt.xlabel('x')
plt.ylabel('u(x,t)')
plt.grid(True)
plt.legend()
plt.show()
```
