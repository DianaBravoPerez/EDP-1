La empresa TIBASA (Fabricante de tinas de baño) tiene asignado un camión especial para el transporte de tinas terminadas. Dicho camión transporta diariamente 5 tinas. El peso de cada tina sigue la distribución triangular de la imagen anexa.

Si la capacidad del camión es de 1 tonelada, ¿cuál es la probabilidad de que el peso de las tinas exceda la capacidad del camión?

- Mínimo: 190 kg
- Moda: 210 kg
- Máximo: 230 kg

La capacidad del camión es de **1 tonelada (1000 kg)**.



```python
import random
import matplotlib.pyplot as plt
```

```python
# Parámetros de la distribución triangular
a = 190   # valor mín
c = 210   # moda
b = 230   # valor máx

# Parámetros de la simulación
n_tinas = 5
capacidad = 1000
n_simulaciones = 10000
```



En cada iteración:

Se generan 5 pesos aleatorios con distribución triangular
Se suman para obtener el peso total del camión
Se verifica si ese peso excede la capacidad

```python
random.seed(42)

veces_excede = 0
pesos_totales = []

for i in range(n_simulaciones):

    # Generar el peso de las 5 tinas y sumarlas
    peso_total = 0
    for j in range(n_tinas):
        peso = random.triangular(a, b, c)
        peso_total = peso_total + peso

    pesos_totales.append(peso_total)

    # Verificar si se excede la capacidad
    if peso_total > capacidad:
        veces_excede = veces_excede + 1

# Calcular la probabilidad estimada
probabilidad = veces_excede / n_simulaciones

print('Núm de simulaciones:           ', n_simulaciones)
print('Veces que se excedió la capacidad:', veces_excede)
print('Probabilidad aprox:            ', round(probabilidad, 4))
```

## Gráfica de resultados

```python
plt.figure(figsize=(8, 5))

plt.hist(pesos_totales, bins=50, color='steelblue', edgecolor='black', label='Peso total simulado')
plt.axvline(x=capacidad, color='red', linestyle='--', linewidth=2, label='Capacidad del camión (1000 kg)')

plt.xlabel('Peso total de las 5 tinas (kg)')
plt.ylabel('Frecuencia')
plt.title('Simulación del peso total diario del camión TIBASA')
plt.legend()
plt.tight_layout()
plt.show()

print(f'La probabilidad de que el peso exceda 1 tonelada es aproximadamente: {probabilidad:.4f} ({probabilidad*100:.2f}%)')
```
