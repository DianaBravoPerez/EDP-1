**RESOLVER LA ECUACIÓN **

\begin{align*}
\frac{\partial^2 u}{\partial t^2} &= c^2 \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right), \quad 0 < x < 4,\ 0 < y < 2,\ t > 0 \\
u(0, y, t) &= u(4, y, t) = u(x, 0, t) = u(x, 2, t) = 0, \quad \text{(condiciones de frontera)} \\
u(x, y, 0) &= 0.1(4x - x^2)(2y - y^2), \quad \text{(condición inicial)} \\
\left. \frac{\partial u}{\partial t} \right|_{t=0} &= 0
\end{align*}

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

# Parámetros del problema
a, b = 1.0, 1.0     # dimensiones del rectángulo
c = 1.0             # velocidad de propagación
N = 10              # número de términos en x
M = 10              # número de términos en y

# Dominio espacial y temporal
Nx, Ny = 50, 50
x = np.linspace(0, a, Nx)
y = np.linspace(0, b, Ny)
X, Y = np.meshgrid(x, y)

# Tiempo
tmax = 5
Nt = 200
t = np.linspace(0, tmax, Nt)

# Condiciones iniciales
def f(x, y):
    return np.sin(np.pi * x / a) * np.sin(np.pi * y / b)  # ejemplo

def g(x, y):
    return 0  # condición inicial de velocidad nula

# Cálculo de coeficientes A_{mn} y B_{mn}
def compute_coefficients(f, g, N, M):
    A = np.zeros((M, N))
    B = np.zeros((M, N))
    for m in range(1, M+1):
        for n in range(1, N+1):
            # Integral doble (cuadratura simple)
            integrand_f = f(X, Y) * np.sin(n*np.pi*X/a) * np.sin(m*np.pi*Y/b)
            integrand_g = g(X, Y) * np.sin(n*np.pi*X/a) * np.sin(m*np.pi*Y/b)
            A[m-1, n-1] = (4/(a*b)) * np.trapz(np.trapz(integrand_f, x, axis=1), y)
            if np.any(g(X,Y)):
                omega_mn = np.pi * np.sqrt((n/a)**2 + (m/b)**2)
                B[m-1, n-1] = (4/(a*b)) * np.trapz(np.trapz(integrand_g, x, axis=1), y) / (c * omega_mn)
    return A, B

A, B = compute_coefficients(f, g, N, M)

# Solución u(x, y, t)
def u_series(x, y, t, A, B):
    u = np.zeros_like(x)
    for m in range(1, M+1):
        for n in range(1, N+1):
            omega_mn = np.pi * np.sqrt((n/a)**2 + (m/b)**2)
            term = (A[m-1, n-1] * np.cos(c * omega_mn * t) +
                    B[m-1, n-1] * np.sin(c * omega_mn * t))
            u += term * np.sin(n*np.pi*x/a) * np.sin(m*np.pi*y/b)
    return u

# Animación
fig = plt.figure()
ax = plt.axes(projection='3d')
surf = [ax.plot_surface(X, Y, f(X, Y), cmap='viridis')]

def animate(i):
    ax.clear()
    Z = u_series(X, Y, t[i], A, B)
    ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_zlim(-1, 1)
    ax.set_title(f"t = {t[i]:.2f}")

ani = animation.FuncAnimation(fig, animate, frames=Nt, interval=50)
plt.show()
```
