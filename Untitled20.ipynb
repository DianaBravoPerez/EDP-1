Queremos verificar que la ecuación:  
$
f''(x) + 2f'(x) + \lambda f(x) = 0, \quad 0 < x < 1
$  
con condiciones de frontera  
$
f(0) = 0, \quad f(1) = 0
$  
es un problema de Sturm–Liouville. Reescribimos la ecuación en forma autoadjunta:

Multiplicamos por el factor integrante $ \mu(x) = e^{2x} $:

$
e^{2x} f''(x) + 2e^{2x} f'(x) + \lambda e^{2x} f(x) = 0
$

El lado izquierdo es:

$
\frac{d}{dx} \left( e^{2x} f'(x) \right) + \lambda e^{2x} f(x) = 0
$

Ahora está en la forma estándar de Sturm–Liouville:

$
\frac{d}{dx} \left[ p(x) f'(x) \right] + \left[ \lambda w(x) - q(x) \right] f(x) = 0
$

con:  
- $ p(x) = e^{2x} $  
- $ w(x) = e^{2x} $  
- $ q(x) = 0 $

Así que **sí es un problema de Sturm–Liouville**.

La solución general de la ecuación homogénea es:

$
f(x) = e^{-x}(C_1 \sin(\sqrt{\lambda} x) + C_2 \cos(\sqrt{\lambda} x))
$

Aplicando las condiciones de frontera

$f''(x) + 2f'(x) + \lambda f(x) = 0, \quad 0 < x < 1
\]
\[
f(0) = 0, \quad f(1) = 0 $

se obtienen los valores propios y funciones propias:

- $ \lambda_n = n^2 \pi^2 $  
- $ f_n(x) = e^{-x} \sin(n \pi x) $, $ n = 1, 2, 3, \dots $



Sea $u(x)$ una función en el intervalo $[a, b]$ y sea $\{f_n(x)\}$ un conjunto de funciones propias ortogonales respecto a una función peso $w(x)$, del problema de Sturm–Liouville. Entonces, la expansión en serie de Fourier generalizada de $u(x)$ es:
$
\[
u(x) = \sum_{n=1}^{\infty} a_n f_n(x)
\]
$
donde los coeficientes $a_n$ están dados por:
$
\[
a_n = \frac{\int_a^b u(x) f_n(x) w(x)\, dx}{\int_a^b f_n^2(x) w(x)\, dx}
\]
$
Para el problema específico:

- Intervalo: $[0, 1]$
- Funciones propias: $f_n(x) = e^{-x} \sin(n \pi x)$
- Función peso: $w(x) = e^{2x}$

La serie queda:
$
\[
u(x) = \sum_{n=1}^{\infty} a_n e^{-x} \sin(n \pi x)
\]
$
con
$
\[
a_n = \frac{\int_0^1 u(x) e^{-x} \sin(n \pi x) e^{2x} dx}{\int_0^1 \left( e^{-x} \sin(n \pi x) \right)^2 e^{2x} dx}
= \frac{\int_0^1 u(x) \sin(n \pi x) e^{x} dx}{\int_0^1 \sin^2(n \pi x) e^{x} dx}
\]
$

**(b) Usar las eigenfunciones de este problema para expandir en series de Fourier generalizadas la función:
$
\[
u(x) = e^{-x}, \quad 0 \leq x \leq 1
\]
$ ** ****

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Variables simbólicas
x, n = sp.symbols('x n')
u = sp.exp(-x)  # Función a expandir
phi_n = sp.exp(-x) * sp.sin(n * sp.pi * x)
w = sp.exp(2*x)

# Calcular coeficiente a_n
numerador = sp.integrate(u * phi_n * w, (x, 0, 1))
denominador = sp.integrate(phi_n**2 * w, (x, 0, 1))
a_n = sp.simplify(numerador / denominador)

# Convertir a función numérica
a_n_func = sp.lambdify(n, a_n, "numpy")

# Definir funciones propias numéricas
def phi(n_val, x_vals):
    return np.exp(-x_vals) * np.sin(n_val * np.pi * x_vals)

# Aproximar la función con N términos
def fourier_expansion(x_vals, N=10):
    s = np.zeros_like(x_vals)
    for k in range(1, N+1):
        s += a_n_func(k) * phi(k, x_vals)
    return s

# Graficar
x_vals = np.linspace(0, 1, 400)
u_vals = np.exp(-x_vals)
approx_vals = fourier_expansion(x_vals, N=10)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, u_vals, label=r'$u(x) = e^{-x}$', linewidth=2)
plt.plot(x_vals, approx_vals, label='Aproximación de Fourier (N=10)', linestyle='--')
plt.title("Expansión en Serie de Fourier Generalizada")
plt.xlabel("x")
plt.ylabel("u(x)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
