### Ecuación de onda -Método explícito

Queremos aproximar la solución de la ecuación de onda

$u_{tt} = 4\,u_{xx}$ en $0 < x < 1$, $0 < t$

con $u(0,t) = u(1,t) = 0$ y condiciones iniciales

$u(x,0)=\sin(\pi x)$, $u_t(x,0)=0$


```python
import sympy as sp            # Cálculo simbólico y manejo de matrices
import numpy as np            # Arreglos numéricos y trabajo con la malla
import matplotlib.pyplot as plt   # Gráficas 2D
from mpl_toolkits.mplot3d import Axes3D   # Soporte para gráficas 3D
from matplotlib import cm                 # Mapas de color
from tabulate import tabulate             # Para tablas (se usará en otros ejercicios)

# Función inicial u(x,0) = sin(pi x)
def f(x):
    """Condición inicial u(x,0) = sin(pi*x)."""
    return np.sin(np.pi * x)
```

Para escribir el esquema en forma matricial usamos solo los nodos internos  
Sea \(n_x = 9\) el número de nodos internos (\(x = 0.1,\dots,0.9\))  

Con \(r = 1\) la matriz tridiagonal queda

\[
A = \begin{pmatrix}
2-2r & r      &        &        & 0 \\
r    & 2-2r   & r      &        &   \\
     & \ddots & \ddots & \ddots &   \\
     &        & r      & 2-2r   & r \\
0    &        &        & r      & 2-2r
\end{pmatrix},
\]

y para \(r = 1\) se tiene \(2-2r = 0\).  
El esquema matricial se puede escribir como

\[
\mathbf{u}^{n+1} = A\,\mathbf{u}^{n} - \mathbf{u}^{n-1}.
\]

```python
def tridiagonal(n, a, b):
    """
    Construye una matriz tridiagonal de tamaño n×n con:
    - 'a' en la diagonal principal
    - 'b' en las diagonales superior e inferior
    Usamos sympy.Matrix para poder multiplicar simbólicamente.
    """
    return sp.Matrix([
        [a if i == j else b if abs(i - j) == 1 else 0 for j in range(n)]
        for i in range(n)
    ])

# Parámetros del problema
h = 0.1       # Paso en x, nodos internos: 0.1, 0.2, ..., 0.9
k = 0.05      # Paso de tiempo (k)
r = 1         # r = (c*k/h)^2 = 1, de acuerdo con el enunciado

nx_internos = 9              # Número de nodos internos en x
t_max = 0.5                  # Tiempo máximo que queremos (t = 0.50)
nt = int(t_max / k)          # Número de pasos de tiempo (0.0 ... 0.5 cada 0.05) → 10
j_max = nt + 1               # Niveles de tiempo totales (incluyendo n=0)

# Construimos la matriz A del esquema explícito
A = tridiagonal(nx_internos, 2 - 2*r, r)

# Vector de condición inicial en los nodos internos x = 0.1,...,0.9
u0 = sp.Matrix([f(h * (i + 1)) for i in range(nx_internos)])

# Primer paso en tiempo:
# Como la velocidad inicial es cero, en clase se aproximó u^1 ≈ u^0
u1 = u0.copy()


def onda_explicita(A, u0, u1, j):
    """
    Resuelve la ecuación de onda con el esquema explícito:
        u^{n+1} = A*u^n - u^{n-1}
    Parámetros:
    - A  : matriz tridiagonal del esquema
    - u0 : vector columna con u(x,0) en los nodos internos
    - u1 : vector columna con u(x,k) (primer nivel de tiempo)
    - j  : número de niveles de tiempo a generar

    Regresa:
    - Matriz (j × nx) donde cada fila corresponde a un tiempo n*k
    """
    # Aseguramos que u0 y u1 sean vectores columna de sympy
    u0 = sp.Matrix(u0)
    u1 = sp.Matrix(u1)

    # Lista que guarda las soluciones en cada tiempo
    S = [u0, u1]

    # Bucle en tiempo: genera u^2, u^3, ..., u^{j-1}
    for n in range(2, j):
        sol = A * S[n - 1] - S[n - 2]   # Fórmula del esquema explícito
        S.append(sol)

    # Regresamos una matriz donde cada fila es (u^n)^T
    return sp.Matrix([s.T for s in S])
```

## Malla \((x,t)\) y solución numérica

Con los valores de \(h\) y \(k\) la malla queda

\[
x_i = i h,\quad i = 1,\dots,9,\qquad
t_n = n k,\quad n = 0,\dots,10,
\]

de modo que el último tiempo es \(t_{10} = 0.5\).  

Calculamos la matriz \(U\) donde cada fila corresponde al vector

\[
\mathbf{u}^n = \big(u(x_1,t_n),\dots,u(x_9,t_n)\big)
\]

y después la convertimos a `numpy` para poder graficar en 2D y 3D

```python
# Calculamos la solución numérica en todos los tiempos hasta t = 0.5
U = onda_explicita(A, u0, u1, j_max)

# Convertimos la matriz de sympy a un arreglo de numpy de tipo float
U_np = np.array(U.tolist(), dtype=float)

# Vector de nodos internos en x y vector de tiempos
x = np.linspace(h, 1 - h, nx_internos)  # 0.1, 0.2, ..., 0.9
t = np.linspace(0, t_max, j_max)       # 0.00, 0.05, ..., 0.50

# Malla para graficar (X: posiciones, T: tiempos)
X, T = np.meshgrid(x, t)

# Verificación rápida de dimensiones
print("Dimensión de U_np:", U_np.shape)   # Debe ser (j_max, nx_internos)
```

## Gráfica 2D de la solución

Para la gráfica 2D usamos un mapa de contorno de \(u(x,t)\)  
En el plano \((x,t)\) se dibujan niveles de la solución numérica, lo que ayuda a ver cómo se propaga la onda  

La figura representa \(u(x,t)\) en la malla hasta \(t = 0.5\)

```python
plt.figure(figsize=(8, 6))

# Mapa de contorno relleno de la solución numérica
contour = plt.contourf(X, T, U_np, levels=25, cmap=cm.viridis)

plt.colorbar(contour, label="u(x,t)")   # Barra de colores para los valores de u
plt.title("Solución numérica (método explícito) de la ecuación de onda")
plt.xlabel("x")
plt.ylabel("t")
plt.tight_layout()
plt.show()
```

## Gráfica 3D de la solución

Ahora representamos la misma solución como una superficie en 3D  

En el eje horizontal aparecen \(x\) y \(t\), y en el eje vertical el valor de \(u(x,t)\)  
Esto permite ver las oscilaciones de la onda a lo largo del tiempo hasta \(t = 0.5\)

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Superficie 3D de la solución numérica
surf = ax.plot_surface(X, T, U_np, cmap=cm.viridis, edgecolor='k')

ax.set_title("Ecuación de onda: solución numérica con método explícito")
ax.set_xlabel("x")
ax.set_ylabel("t")
ax.set_zlabel("u(x,t)")

fig.colorbar(surf, shrink=0.5, aspect=5, label="u(x,t)")
plt.tight_layout()
plt.show()
```

## Ejercicio 2: comparación con la Tabla 12.6

La Tabla 12.6 da los valores numéricos \(w_{i,20}\) para el mismo problema  

Con \(k = 0.05\) el nivel de tiempo \(n = 20\) corresponde a

\[
t_{20} = 20 k = 20 \cdot 0.05 = 1.0 .
\]

La tabla incluye los once nodos \(x_i = 0, 0.1, \dots, 1.0\) con

\[
w_{i,20} \approx u(x_i,1).
\]

Nuestro método explícito solo calcula los nodos internos, así que para comparar
añadimos las fronteras \(u(0,t) = u(1,t) = 0\)  
Luego graficamos \(w_{i,20}\) y la solución numérica en \(t = 1\), y
calculamos el error punto a punto

\[
e_i = u_{\text{num}}(x_i,1) - w_{i,20}.
\]

```python
# --- Ejercicio 2: solución numérica hasta t = 1.0 ---

# Tiempo máximo para este ejercicio
t_max_E2 = 1.0
nt_E2 = int(t_max_E2 / k)      # número de pasos: 1.0 / 0.05 = 20
j_E2 = nt_E2 + 1               # niveles de tiempo (n = 0,...,20) → 21 filas

# Volvemos a usar A, u0 y u1 definidos antes
U_E2 = onda_explicita(A, u0, u1, j_E2)

# Pasamos a numpy
U_E2_np = np.array(U_E2.tolist(), dtype=float)

# Índice de tiempo n = 20 (t = 1.0)
n_20 = 20
u_num_interno = U_E2_np[n_20, :]   # valores en x = 0.1,...,0.9

# Agregamos las condiciones de frontera (0 en x=0 y x=1)
u_num_completo = np.concatenate(([0.0], u_num_interno, [0.0]))

# Vector de nodos correspondientes a la tabla (incluye fronteras)
x_tabla = np.linspace(0.0, 1.0, 11)
```

## Valores de la Tabla 12.6

Tomamos de la Tabla 12.6 los valores \(w_{i,20}\) para \(x_i = 0, 0.1,\dots,1\):

\[
\{w_{i,20}\}_{i=0}^{10} =
(0,\;0.3090,\;0.5878,\;0.8090,\;0.9511,\;1.0,\;0.9511,\;0.8090,\;0.5878,\;0.3090,\;0).
\]

Estos son los valores de referencia con los que vamos a comparar el esquema explícito

```python
# Valores w_{i,20} de la Tabla 12.6 (redondeados a 10 decimales)
w_tabla = np.array([
    0.0000000000,
    0.3090169944,
    0.5877852523,
    0.8090169944,
    0.9510565163,
    1.0000000000,
    0.9510565163,
    0.8090169944,
    0.5877852523,
    0.3090169944,
    0.0000000000
])

# Error punto a punto
error = u_num_completo - w_tabla
error_abs = np.abs(error)

# Medidas globales de error
max_error = np.max(error_abs)
rmse = np.sqrt(np.mean(error**2))

print(f"Error máximo absoluto: {max_error:.3e}")
print(f"RMSE: {rmse:.3e}\n")

# Tabla bonita para la notebook
tabla_resumen = np.column_stack([x_tabla, w_tabla, u_num_completo, error_abs])
headers = ["x_i", "w_{i,20} (Tabla)", "u_num(x_i,1)", "|error|"]

print(tabulate(tabla_resumen, headers=headers, floatfmt=".10f"))
```

## Gráfica de comparación y error

Primero dibujamos en la misma figura los valores de la Tabla 12.6 y los del método explícito:

\[
x_i \mapsto w_{i,20},\qquad
x_i \mapsto u_{\text{num}}(x_i,1).
\]

Después graficamos el error absoluto \(|e_i|\) vs \(x_i\) para ver en qué puntos
se separan más  

Con esto se puede comentar si la aproximación numérica replica bien los valores de la tabla

```python
# --- Gráfica 1: comparación numérico vs tabla ---

plt.figure(figsize=(8, 5))
plt.plot(x_tabla, w_tabla, "o-", label="Tabla 12.6 ($w_{i,20}$)")
plt.plot(x_tabla, u_num_completo, "s--", label="Método explícito (t=1.0)")
plt.xlabel("x")
plt.ylabel("u(x,1)")
plt.title("Comparación de $w_{i,20}$ con la solución numérica")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# --- Gráfica 2: error absoluto ---

plt.figure(figsize=(8, 5))
plt.stem(x_tabla, error_abs, basefmt=" ")
plt.xlabel("x")
plt.ylabel(r"$|u_{\text{num}}(x,1) - w_{i,20}|$")
plt.title("Error absoluto entre la tabla y el método explícito")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
