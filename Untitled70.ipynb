```python
import numpy as np
import matplotlib.pyplot as plt
import random as r

def test_function(x):
    return np.sin(x)

def mc(x_1, x_2, f, N):
    x = np.linspace(x_1, x_2, 100)
    y1 = 0
    y2 = max(f(x))

    exitos = []
    xs = []
    yx = []
    check = []

    for i in range(N):
        x = r.uniform(x_1, x_2)
        xs.append(x)
        y = r.uniform(y1, y2)
        yx.append(y)

        if abs(y) > abs(f(x)) or y < 0:
            check.append(0)
        else:
            check.append(1)

    estimacion = (np.pi / 2) * np.mean(check)
    return estimacion, xs, yx  # ← ahora retorna los 3

# Llamada
estimacion, e, f = mc(0, np.pi/2, test_function, 1000)

# Gráfica
plt.plot(np.linspace(0, np.pi/2, 100), np.sin(np.linspace(0, np.pi/2, 100)))
plt.scatter(e, f, color="red", s=5)
plt.title(f"Estimación I ≈ {estimacion:.4f}")
plt.show()
```
