```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,0.1)
y = x*x

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafica', color='red')
plt.show()


```

Graficar: $$ y = x^2 $$
