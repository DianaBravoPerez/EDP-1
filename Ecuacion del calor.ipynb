# An치lisis de una soluci칩n de la ecuaci칩n del calor

b

Este problema modela la transferencia de calor en una barra ideal delgada, con $k=1$.

```python
from pylab import *
```

Para empezar, graficamos la condici칩n inicial.

```python
def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i]
    return y
```

```python
x = linspace(0,1,1000)
f = recta(x)
plot(x,f,label='$u(x,0)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Condici칩n inicial (u(x,0)=f(x))")
xlim([-0.01,1.01]),ylim([-0.01,1.01]),grid(),legend()
show()
```

Usamos una serie de Fourier para aproximar la condici칩n inicial $u(x,0)=f(x)=x$.

```python
def u_0(x,N=20): # N es el n칰mero de arm칩nicos
    f = zeros(len(x))
    for n in range(1,N):
        f += (-1)**(n+1)*sin(n*pi*x)/(n)
    return (2.0/pi)*f
```

```python
f_n = u_0(x)
plot(x,f_n,label='$f_N(x)$ con $N=20$')
plot(x,f,label='$f(x)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci칩n de la condici칩n inicial")
xlim([-0.01,1.1]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Podemos visualizar la aproximaci칩n para diferentes cantidades de arm칩nicos.

```python
for i in range(1,20,5):
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci칩n de la condici칩n inicial")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Ahora, agregamos la parte temporal para visualizar la soluci칩n completa. Primero, para el momento inicial, $t=0$.

```python
def u_x_t(x,t,N=50):
    f = zeros( (len(x),len(t)) )
    for i in range(len(t)):
        for n in range(1,N):
            f[:,i] += (-1)**(n+1)*sin(n*pi*x)/(n)*exp(-n**2*pi**2*t[i])
    return (2.0/pi)*f
```

```python
x = linspace(0,1,1000)
t = linspace(0,10,1000)
U = u_x_t(x,t)
```

```python
plot(x,U[:,0],label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci칩n de la condici칩n inicial")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Luego, para diferentes tiempos, lo cual nos permite ver c칩mo la soluci칩n evoluciona cuando $t \to \infty$.

```python
plot(x,U[:,0],label='$u(x,t=0)$') # Mejorar
plot(x,U[:,1],label='$u(x,t=1)$')
plot(x,U[:,2],label='$u(x,t=2)$')
plot(x,U[:,3],label='$u(x,t=3)$')
plot(x,U[:,10],label='$u(x,t=10)$')
plot(x,U[:,20],label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$');
plt.title("Evoluci칩n de la soluci칩n")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()
```

Y podemos ver esto en una gr치fica en 3D.

```python
#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0,1.0,100)
t = linspace(0,1.0,100)

X,T = np.meshgrid(x,t)

U = u_x_t(x,t)

plt.xlabel('$t$'); plt.ylabel('$x$');
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```

La soluci칩n de la ecuaci칩n
\begin{equation}
    u_t(x,t) = u_{xx}(x,t), \quad 0 < x < 1, \quad t > 0,
\end{equation}
sujeta a las condiciones de frontera
\begin{equation}
    u_x(0,t) = 0, \quad u_x(1,t) = 0, \quad t > 0,
\end{equation}

\begin{equation}
    u(x,0) = f(x) = x, \quad 0 < x < 1.
\end{equation}
Esta dada en t칠rminos de la serie de Fourier por la expresi칩n:

$
u(x,t) = \frac{1}{2}+ \sum_{n=1}^{\infty} \frac{(2(-1)^{n}-1)}{\pi^{2} n^{2}} e^{-(\pi n)^2 t} \cos(\pi n x).$

```python
from pylab import *

def recta(x):
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i]
    return y

x = linspace(0,1,1000)
f = recta(x)
plot(x,f,label='$u(x,0)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Condici칩n inicial (u(x,0)=f(x))")
xlim([-0.01,1.01]),ylim([-0.01,1.01]),grid(),legend()
show()

def u_0(x, N=20): # N es el n칰mero de arm칩nicos
    f = zeros(len(x))
    for n in range(1, N+1):  # Corregir el rango de n para incluir N t칠rminos
        f += (-1)**(n+1) * sin(n*pi*x) / n
    return (2.0 / pi) * f  # Asegurarse de que la expresi칩n sea correcta

f_n = u_0(x)
plot(x, f_n, label='$f_N(x)$ con $N=20$')
plot(x, f, label='$f(x)=x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci칩n de la condici칩n inicial")
xlim([-0.01,1.1]),ylim([-0.01,1.2]),grid(),legend()
show()

for i in range(1, 20, 5):
    plot(x, u_0(x, i), label='$f_N(x)$ con $N=%d$' % i)
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci칩n de la condici칩n inicial")
xlim([-0.01,1.01]),ylim([-0.01,1.2]),grid(),legend()
show()

def u_x_t(x, t, N=50):
    f = zeros((len(x), len(t)))
    for i in range(len(t)):
        for n in range(1, N+1):  # Corregir el rango de n para incluir N t칠rminos
            f[:, i] += (-1)**(n+1) * sin(n*pi*x) / n * exp(-(n*pi)**2 * t[i])  # Corregir el t칠rmino de la exponencial
    return (2.0 / pi) * f

x = linspace(0, 1, 1000)
t = linspace(0, 10, 1000)
U = u_x_t(x, t)

plot(x, U[:, 0], label='$u(x,t=0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximaci칩n de la condici칩n inicial")
xlim([-0.01,1.01]), ylim([-0.01,1.2]), grid(), legend()
show()

plot(x, U[:, 0], label='$u(x,t=0)$')  # Mejorar
plot(x, U[:, 1], label='$u(x,t=1)$')
plot(x, U[:, 2], label='$u(x,t=2)$')
plot(x, U[:, 3], label='$u(x,t=3)$')
plot(x, U[:, 10], label='$u(x,t=10)$')
plot(x, U[:, 20], label='$u(x,t=20)$')
plt.xlabel('$x$'); plt.ylabel('$u(x,t)$');
plt.title("Evoluci칩n de la soluci칩n")
xlim([-0.01,1.01]), ylim([-0.01,1.2]), grid(), legend()
show()

#%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection="3d")

x = linspace(0, 1.0, 100)
t = linspace(0, 1.0, 100)

X, T = np.meshgrid(x, t)

U = u_x_t(x, t)

plt.xlabel('$t$'); plt.ylabel('$x$');
plt.title("u(x,t)")
ax.plot_surface(X, T, U, cmap="inferno"); #Otros colores:"viridis", "inferno", "plasma"
```
