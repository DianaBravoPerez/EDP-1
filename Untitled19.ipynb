***PROBLEMAS DE STURM-LIOVILLE***



1) $$ X'' + kX = 0,   X(0)=X(1)=0 $$




```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, Eq, dsolve, solve, lambdify, sqrt

# Variables simbólicas
t = symbols('t')
k = symbols('k', positive=True)
C1, C2 = symbols('C1 C2')
X = Function('X')

# Resolver EDO
sol_general = dsolve(Eq(X(t).diff(t, 2) + k * X(t), 0), X(t))
print("Solución general:")
print(sol_general)

# Aplicar X(0) = 0 → C2 = 0
cond1 = sol_general.rhs.subs(t, 0)
C2_val = solve(cond1, C2)[0]

# Sustituir C2 en la solución
sol_sust = sol_general.rhs.subs(C2, C2_val)

# Aplicar X(1) = 0 para encontrar k
cond2 = sol_sust.subs(t, 1)
k_vals = solve(cond2, k)
print("Valores permitidos de k:")
print(k_vals)

# Escoger primer valor válido de k
final_expr = sol_sust.subs(k, k_vals[0])
print("Solución particular:")
print(f"X(t) = {final_expr}")

# Escoger un valor fijo de C1
C1_val = 1  # <-- se puede cambiar este valor

# Convertir a función numérica
X_func = lambdify(t, final_expr.subs(C1, C1_val), 'numpy')
t_vals = np.linspace(0, 1, 400)
y_vals = X_func(t_vals)

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(t_vals, y_vals, label=f'C1 = {C1_val}')
plt.title("Solución particular de la EDO con condiciones de frontera")
plt.xlabel("t")
plt.ylabel("X(t)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

2) $$X'' + kX = 0,   X'(0)=X'(1)=0$$


```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Function, Eq, dsolve, diff, solve, lambdify, sqrt

# Variables simbólicas
t = symbols('t')
k = symbols('k', positive=True)
C1, C2 = symbols('C1 C2')
X = Function('X')

# Resolver EDO
sol_general = dsolve(Eq(X(t).diff(t, 2) + k * X(t), 0), X(t))
print("Solución general:")
print(sol_general)

# Derivada de la solución
sol_deriv = diff(sol_general.rhs, t)
print("\nDerivada de la solución general:")
print(sol_deriv)

# Aplicar condición X'(0) = 0
deriv_0 = sol_deriv.subs(t, 0)
C1_val = solve(deriv_0, C1)[0]
print(f"\nValor de C1 obtenido de X'(0)=0: C1 = {C1_val}")

# Sustituimos C1 en la derivada
sol_deriv_sub = sol_deriv.subs(C1, C1_val)

# Aplicar condición X'(1) = 0
deriv_1 = sol_deriv_sub.subs(t, 1)
k_vals = solve(deriv_1, k)
print("\nValores permitidos de k:")
print(k_vals)

# Sustituimos en la solución original
final_expr = sol_general.rhs.subs({C1: C1_val, k: k_vals[0]})
print("\nSolución particular con condiciones de derivadas:")
print(f"X(t) = {final_expr}")

# Elegimos un valor para C2 y graficamos
C2_val = 1  # <-- Puedes cambiar este valor si quieres

# Convertimos a función numérica
X_func = lambdify(t, final_expr.subs(C2, C2_val), 'numpy')

# Graficar
t_vals = np.linspace(0, 1, 400)
y_vals = X_func(t_vals)

plt.figure(figsize=(8, 5))
plt.plot(t_vals, y_vals, label=f'C2 = {C2_val}')
plt.title("Solución para X'(0)=0 y X'(1)=0")
plt.xlabel("t")
plt.ylabel("X(t)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
