```python
import random
import math
import matplotlib.pyplot as plt
```

## Trayectoría con T = 1 y L = 100

Siguiendo el seudocódigo:
empezamos con $W_0 = 0$ y en cada paso le sumamos $\sqrt{\delta t}\,\xi_i$, donde $\xi_i \sim N(0,1)$

```python
# Parametros
T = 1
L = 100
dt = T / L

# Aplicamos el seudocodigo
W = [0]            # W_0 = 0
for i in range(L):
    xi = random.gauss(0, 1)  # xi ~ N(0,1)
    W.append(W[i] + math.sqrt(dt) * xi)
# Tiempos t_i = i*dt
t = [i * dt for i in range(L + 1)]


#Graficamos 100 trayectorías
plt.figure(figsize=(8,5))
plt.plot(t, W)
plt.title("Trayectoría del movimiento Browniano")
plt.xlabel("t")
plt.ylabel("W(t)")
plt.grid(True)
plt.show()



```

```python
T = 1
L = 100
dt = T / L
t = [i * dt for i in range(L + 1)]

plt.figure(figsize=(8, 5))

# Generamos 100 trayectorias
for k in range(100):
    W = [0]
    for i in range(L):
        xi = random.gauss(0, 1)
        W.append(W[i] + math.sqrt(dt) * xi)
    plt.plot(t, W, linewidth=0.7)

# Funciones f(t) = sqrt(t) y g(t) = -sqrt(t)
f = [math.sqrt(x) for x in t]
g = [-math.sqrt(x) for x in t]
plt.plot(t, f, color="black", linewidth=2, label="f(t) = sqrt(t)")
plt.plot(t, g, color="black", linewidth=2, label="g(t) = -sqrt(t)")

plt.title("100 trayectorias del movimiento Browniano")
plt.xlabel("t")
plt.ylabel("W(t)")
plt.legend()
plt.grid(True)
plt.show()
```
