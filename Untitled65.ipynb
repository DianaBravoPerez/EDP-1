```python
def prueba_rachas(datos, alpha=0.05):
    """
    Aplica la Prueba de Rachas a una secuencia de numeros.

    Parametros
    ----------
    datos : array-like
    alpha : nivel de significancia (default 0.05)
    """
    datos = np.array(datos)
    mediana = np.median(datos)

    # Asignar signos respecto a la mediana
    signos = np.where(datos >= mediana, 1, -1)

    # Contar n1 y n2
    n1 = np.sum(signos == 1)
    n2 = np.sum(signos == -1)

    # Contar rachas (cambia cuando el signo es distinto al anterior)
    R = 1 + np.sum(signos[1:] != signos[:-1])

    # Media y desviacion estandar esperadas
    mu_R = (2 * n1 * n2) / (n1 + n2) + 1
    sigma_R = np.sqrt(
        (2 * n1 * n2 * (2 * n1 * n2 - n1 - n2)) /
        ((n1 + n2)**2 * (n1 + n2 - 1))
    )

    # Estadistico Z
    Z = (R - mu_R) / sigma_R
    Z_critico = stats.norm.ppf(1 - alpha / 2)
    valor_p = 2 * (1 - stats.norm.cdf(abs(Z)))

    print(f"Mediana  : {mediana:.4f}")
    print(f"n1       : {n1}")
    print(f"n2       : {n2}")
    print(f"R        : {R}")
    print(f"mu_R     : {mu_R:.4f}")
    print(f"sigma_R  : {sigma_R:.4f}")
    print(f"Z        : {Z:.4f}")
    print(f"Z critico: +/-{Z_critico:.4f}")
    print(f"Valor-p  : {valor_p:.4f}")

    if abs(Z) <= Z_critico:
        print("Decision : No se rechaza H0. La secuencia es aleatoria.")
    else:
        print("Decision : Se rechaza H0. La secuencia no es aleatoria.")

    return signos, R, Z, Z_critico


signos, R, Z, Z_critico = prueba_rachas(numeros)
```
