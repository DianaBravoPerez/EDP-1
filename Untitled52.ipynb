# **método de colocación**

```python
# Método de colocación para el Ejercicio 5.1
# u'' + u = x,  u(0) = 0,  u'(0) = 2

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# 1) Declaramos las variables simbólicas
x = sp.symbols('x')
a0, a1, a2, a3 = sp.symbols('a0 a1 a2 a3')

# 2) Definimos el polinomio cúbico aproximante
uN = a0 + a1*x + a2*x**2 + a3*x**3

# 3) Condiciones iniciales u(0)=0, u'(0)=2
uN_prime = sp.diff(uN, x)

eq_ci1 = sp.Eq(uN.subs(x, 0), 0)   # u(0) = 0
eq_ci2 = sp.Eq(uN_prime.subs(x, 0), 2)  # u'(0) = 2

# 4) Definimos el residuo epsilon(x) = uN'' + uN - x
uN_second = sp.diff(uN, x, 2)
eps = uN_second + uN - x

# 5) Ecuaciones de colocación: epsilon(1/2)=0 y epsilon(1)=0
eq_col1 = sp.Eq(eps.subs(x, sp.Rational(1, 2)), 0)
eq_col2 = sp.Eq(eps.subs(x, 1), 0)

# 6) Resolvemos el sistema para los coeficientes a0,a1,a2,a3
sol_coef = sp.solve((eq_ci1, eq_ci2, eq_col1, eq_col2),
                    (a0, a1, a2, a3), dict=True)[0]

print("Coeficientes de u_N(x):")
print(sol_coef)

# Sustituimos los coeficientes en uN y simplificamos
uN_aprox = sp.simplify(uN.subs(sol_coef))
print("\nAproximación por colocación u_N(x) =")
print(uN_aprox)

# 7) Obtenemos la solución exacta con dsolve para comparar
u = sp.Function('u')
ode = sp.Eq(sp.diff(u(x), x, 2) + u(x), x)
sol_exact = sp.dsolve(ode, ics={u(0): 0,
                                sp.diff(u(x), x).subs(x, 0): 2})
u_exact = sp.simplify(sol_exact.rhs)

print("\nSolución exacta u(x) =")
print(u_exact)

# 8) Definimos el error simbólico E(x) = u(x) - u_N(x)
E = sp.simplify(u_exact - uN_aprox)
print("\nError simbólico E(x) = u(x) - u_N(x) =")
print(E)

# 9) Versión numérica para graficar y medir el error
uN_num = sp.lambdify(x, uN_aprox, "numpy")
u_exact_num = sp.lambdify(x, u_exact, "numpy")
E_num = sp.lambdify(x, E, "numpy")

# Malla de puntos en [0,1]
xx = np.linspace(0, 1, 200)
u_exact_vals = u_exact_num(xx)
uN_vals = uN_num(xx)
err_vals = u_exact_vals - uN_vals

# Error máximo en el intervalo [0,1]
err_max = np.max(np.abs(err_vals))
print("\nError máximo en [0,1] ≈", float(err_max))

# 10) Gráfica de solución exacta vs aproximación
plt.figure(figsize=(8, 5))
plt.plot(xx, u_exact_vals, label='u(x) exacta')
plt.plot(xx, uN_vals, '--', label='u_N(x) colocación')
plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Solución exacta vs aproximación por colocación')
plt.grid(True)
plt.legend()

# 11) Gráfica del error E(x)
plt.figure(figsize=(8, 5))
plt.plot(xx, err_vals, label='E(x) = u(x) - u_N(x)')
plt.xlabel('x')
plt.ylabel('Error')
plt.title('Error de la aproximación por colocación')
plt.grid(True)
plt.legend()

plt.show()
```
