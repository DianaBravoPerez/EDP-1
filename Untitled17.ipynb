```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definir la funcin por partes
def f(x):
    return np.piecewise(x, [(-2 <= x) & (x < 0), (0 <= x) & (x <= 2)], [0, lambda x: 2 - x])

# Parmetros
L = 2  # Perodo de la funcin es 4 (de -2 a 2, simtrico)
T = 4  # Perodo total
harmonics = [5, 10, 20, 50]  # Diferentes valores de N

# Definir variable simblica para clculo de coeficientes
x = sp.Symbol('x')

# Funcin por partes en trminos simblicos
f_sym = sp.Piecewise((0, (-2 <= x) & (x < 0)), (2 - x, (0 <= x) & (x <= 2)))

# Clculo de los coeficientes de Fourier
a0 = (1/T) * sp.integrate(f_sym, (x, -2, 2))
an = lambda n: (2/T) * sp.integrate(f_sym * sp.cos(n * sp.pi * x / L), (x, -2, 2))
bn = lambda n: (2/T) * sp.integrate(f_sym * sp.sin(n * sp.pi * x / L), (x, -2, 2))

# Convertimos los coeficientes a valores numricos
a0_val = float(a0)

# Funcin aproximada con la Serie de Fourier
def fourier_series(x, N):
    an_vals = [float(an(n)) for n in range(1, N + 1)]
    bn_vals = [float(bn(n)) for n in range(1, N + 1)]
    result = a0_val / 2
    for n in range(1, N + 1):
        result += an_vals[n - 1] * np.cos(n * np.pi * x / L) + bn_vals[n - 1] * np.sin(n * np.pi * x / L)
    return result

# Rango de valores para graficar
x_vals = np.linspace(-4, 4, 400)
f_vals = f(x_vals)

# Crear figuras separadas
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Graficar para cada nmero de armnicos
for i, N in enumerate(harmonics):
    fs_vals = fourier_series(x_vals, N)
    ax = axs[i // 2, i % 2]  # Para acomodar en la cuadrcula de 2x2
    ax.plot(x_vals, fs_vals, 'b', label=f'N = {N}')
    ax.plot(x_vals, f_vals, 'r--', label='Funcin original')  # Lnea punteada para la original
    ax.set_title(f'Aproximacin con {N} armnicos')
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
    ax.legend()
    ax.grid()

# Ajustar diseo
plt.tight_layout()
plt.show()
```
